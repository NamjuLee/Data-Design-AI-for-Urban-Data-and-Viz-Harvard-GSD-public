```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def renderLine(x0, y0, x1, y1):
    canvas = np.zeros((200, 200, 3), np.uint8)
    imgLine = cv2.line(canvas, (x0, y0), (x1, y1), (255, 255 ,255), 2 )
    plt.imshow(imgLine)
```

```python
line = [ 10, 10, 100, 100]
print(line)
```

```python
renderLine(line[0], line[1], line[2] ,line[3])
```

```python
p0 = [10, 100]
p1 = [100, 10]

line = [p0, p1]
print(line)
```

```python
renderLine(line[0][0], line[0][1], line[1][0] ,line[1][1])
```

```python
def renderLineSeg(array):
    canvas = np.zeros((200, 200, 3), np.uint8)

    for i in range(0, len(array) - 2 , 2):
        print(i)
        imgLine = cv2.line(canvas, (array[i][0], array[i][1]), (array[i+1][0], array[i+1][1]), (255, 255 ,255), 2 )
    plt.imshow(imgLine)


lns = []
for y in range(10, 200, 10):
    for x in range(10, 200, 10):

        if (x + y) % 2 == 0:
            lns.append([x, y] )

renderLineSeg(lns)
```

```python
def renderLineClass(array):
    canvas = np.zeros((200, 200, 3), np.uint8)
    imgLine = cv2.line(canvas, array[0], array[1], (255, 255 ,255), 2 )
    plt.imshow(imgLine)

class Point():
    def __init__(self, _x, _y,):
        self.x = _x
        self.y = _y

class Line():
    def __init__(self, p0, p1):
        self.p0 = p0
        self.p1 = p1
    def toRender(self):
        return [(self.p0.x, self.p0.y), (self.p1.x, self.p1.y)] 
    # def toString(self):
        #
    # def isIntersect(self, other):
        # https://www.geeksforgeeks.org/point-of-intersection-of-two-lines-formula
```

```python
p0 = Point(50, 80)
p1 = Point(150, 90)

ln = Line(p0, p1)
print(ln.toRender())

renderLineClass(ln.toRender())
```

Polyline

```python
def renderPolyline(array):
    canvas = np.zeros((200, 200, 3), np.uint8)
    for i in range(0, len(array) - 2, 2):
        imgLine = cv2.line(canvas, (array[i], array[i+1]), (array[i+2], array[i+3]), (255, 255 ,255), 2 )
    plt.imshow(imgLine)
```

```python
pline = [ 50, 50, 50, 100, 100, 50, 100, 100 ]

renderPolyline(pline)
```

Poygon

```python
def renderPolygon(pts):
    canvas = np.zeros((200, 200, 3), np.uint8)
    pts = np.array(pts, np.int32)
    cv2.fillPoly(canvas, [pts], (0,255,255))
    plt.imshow(canvas)
```

```python
pline = [ [50, 50], [50, 100], [100, 50], [100, 100] ]
renderPolygon(pline)
```

```python
pGon =[[80,50],[100, 100],[140, 50]]

renderPolygon(pGon)
```
