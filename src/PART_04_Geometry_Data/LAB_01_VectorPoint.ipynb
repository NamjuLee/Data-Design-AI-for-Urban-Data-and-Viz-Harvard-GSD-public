```python
def renderPoint(x, y, r = 2):
    canvas = np.zeros((200, 200, 3), np.uint8)
    plt.imshow(canvas)
    result = cv2.circle(canvas, (x, y),r, (255, 255, 255), 2)
    plt.imshow(result)
```

```python
renderPoint(myVector['x'], myVector['y'])
```

```python
def renderPoints(pts, r = 2):
    canvas = np.zeros((200, 200, 3), np.uint8)
    plt.imshow(canvas)
    for p in pts:
        result = cv2.circle(canvas, (p[0], p[1]),r, (255, 255, 255), 2)
    plt.imshow(result)
    return result;
```

```python
renderPoints(pts)
```

```python
class Vector3():
    listPoint = []

    def __init__(self, _x, _y, _z):
        self.x = _x
        self.y = _y
        self.z = _z
        self.id = len(self.listPoint)
        self.listPoint.append(self)

    def toString(self):
        temp = 'x :' + str(self.x) + ', y :' + str(self.y) + ', z :' + str(self.z)
        # print(temp)
        return temp
    def add(self, v):
        theX = self.x + v.x
        theY = self.y + v.y
        theZ = self.z + v.z
        self.x += v.x
        self.y += v.y
        self.z += v.z
        return Vector3(theX, theY, theZ)
    def sub(self, v):
        theX = self.x + v.x
        theY = self.y + v.y
        theZ = self.z + v.z
        self.x -= v.x
        self.y -= v.y
        self.z -= v.z
        return Vector3(theX, theY, theZ)
```

```python
myV0 = Vector3(10, 20, 30)
myV0.toString()
```

```python
myV1 = Vector3(5, 10, 15)
myV1.toString()
```
