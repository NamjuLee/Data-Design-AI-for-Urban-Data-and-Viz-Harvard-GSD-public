# Color

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def drawingColor(c):
    canvas = np.zeros((50, 50, 3), np.uint8)
    imgRect = cv2.circle(canvas, (25, 25),25, c, 50)
    plt.imshow(imgRect)
```

```python
color = [255, 100, 155, 1.0]

drawingColor((color[0], color[1], color[2]))
```

```python
c0 = (150, 100, 15)
drawingColor(c0)
```

```python
c1 = (50, 70, 150)
drawingColor(c1)
```

```python
def Addition(vA, vB, blend):
    return vA * (1.0 - blend) + blend * (vA + vB) * 0.5

def Subtraction(vA, vB, blend):
    return vA * (1.0 - blend) + blend * (vA - vB) * 2

def Multiply(vA, vB, blend):
    return vA * (1.0 - blend) + blend * (vA * vB) * 0.5

def Screen(vA, vB, blend):
    return vA * (1.0 - blend) + blend * (1 - (1 - vA) * (1 - vB))
```

```python
c3 = []
for i in range(len(c1)):
    print(c0[i], c1[i])
    col = Addition(c0[i] /255, c1[i] / 255, 1)
    c3.append(col)
print(c3)
drawingColor((c3[0] * 255,c3[1] * 255,c3[2] * 255))
```

```python
c3 = []
for i in range(len(c1)):
    print(c0[i], c1[i])
    col = Subtraction(c0[i] / 255, c1[i] / 255, 1)
    c3.append(col)
print(c3)
drawingColor((c3[0] * 255 ,c3[1] * 255,c3[2] * 255))
```

```python
c3 = []
for i in range(len(c1)):
    print(c0[i], c1[i])
    col = Multiply(c0[i] / 255, c1[i] / 255 , 1)
    c3.append(col)
print(c3)
drawingColor((c3[0]* 255, c3[1] * 255, c3[2] * 255))
```

```python
c3 = []
for i in range(len(c1)):
    print(c0[i], c1[i])
    col = Screen(c0[i] / 255 , c1[i] / 255, 1)
    c3.append(col)
print(c3)
drawingColor((c3[0] * 255 ,c3[1] * 255,c3[2] * 255))
```

```python
import math

def distance2(x0, y0, z0, x1, y1, z1):
    return ((x1 - x0) * (x1 - x0)) + ((y1 - y0) * (y1 - y0)) + ((z1 - z0) * (z1 - z0))
def distance(x0, y0, z0, x1, y1, z1):
    return math.sqrt(distance2(x0, y0, z0, x1, y1, z1))

def dis(x0, y0, z0, x1, y1, z1):
    return math.sqrt((x0 - x1) * (x0 - x1) + (y0- y1) * (y0- y1) + (z0 - z1) * (z0 - z1));


def rgb_to_hex(r, g, b):
    return '%02x%02x%02x' % (r,g,b)
def hex_to_rgb(value):
    value = value.lstrip('#')
    lv = len(value)
    return [int(value[i:i+lv//3], 16) for i in range(0, lv, lv//3)]
```

```python
print(hex_to_rgb('#ff0000'))
print(hex_to_rgb('#00ffff'))
```

```python
print(rgb_to_hex(255, 0, 0))
print(rgb_to_hex(0, 255, 255))
```

```python
red = '#9ff32f'
cyan = '#00ffff'

redRGB = hex_to_rgb(red)
cyanRGB = hex_to_rgb(cyan)

print(dis(redRGB[0], redRGB[1], redRGB[2], cyanRGB[0], cyanRGB[1], cyanRGB[2]))
```

```python
drawingColor((redRGB[0], redRGB[1], redRGB[2]))
```
