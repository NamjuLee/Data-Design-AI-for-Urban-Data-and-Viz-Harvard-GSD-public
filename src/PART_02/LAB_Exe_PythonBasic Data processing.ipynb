5. Reverse String data

```python
newTextArray = [];
txt = 'helloWorld';
textOut = ''

for i in txt:
  newTextArray.append(i)

for i in range( int(len(newTextArray) / 2)):
  temp = newTextArray[i];
  newTextArray[i] = newTextArray[len(newTextArray) - i - 1];
  newTextArray[len(newTextArray) - i - 1] = temp;

for i in newTextArray:
  textOut+= i

print(textOut)
```
