```python
import glob
import os
import PIL
import PIL.Image
import cv2
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
PATH_SAVED = "dataProcessed"
files = glob.glob(PATH_SAVED + "/*.*")

im = cv2.imread (files[3], cv2.IMREAD_COLOR)
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)

array = tf.keras.preprocessing.image.img_to_array(im)
plt.imshow(array/255)
```

```python
PATH_SAVED = "dataProcessed"
files = glob.glob(PATH_SAVED + "/*.*")
classList = []

for f in files:
    type = f.split('\\')[-1].split('_')[0]
    if type not in classList:
        classList.append(type)
print(classList)

np.random.seed(222)
np.random.shuffle(files)

data = []
target = []
for f in files:
    type = f.split('\\')[-1].split('_')[0]
    if (type in classList):
        id = classList.index(type)
        target.append(id)

        im = cv2.imread (f, cv2.IMREAD_COLOR)
        im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
        array = tf.keras.preprocessing.image.img_to_array(im)
        # array.resize(DIM[0]))
        data.append(array)
        # plt.imshow(array/255)
        # break
    # print(data[0])

dataArray = np.array(data)
# dataArray = tf.expand_dims(dataArray, axis=-1)
```

```python
from tensorflow.keras.utils import to_categorical
targetOnehot = to_categorical(target)
len(targetOnehot[0])
```

```python
# normalization
dataArrayNormalized = dataArray / 255.
print(dataArrayNormalized.shape)
print(dataArrayNormalized)
```

```python
trainX = dataArrayNormalized[:-50]
testX = dataArrayNormalized[-50:-25]
valX = dataArrayNormalized[-25:]

trainY = targetOnehot[:-50]
testY = targetOnehot[-50:-25]
valY = targetOnehot[-25:]

NUM_CLASS = len(targetOnehot[0])
NUM_CLASS
```

```python
targetOnehot
```

```python
print(dataArrayNormalized.shape)
print(trainX.shape)
print(testX.shape)
print(valY.shape)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(100, 100, 3)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.25),
  tf.keras.layers.Dense(64, activation=tf.nn.relu),
  tf.keras.layers.Dense(NUM_CLASS, activation=tf.nn.softmax)
])

model.compile(loss='categorical_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model)
```

```python
      
hist = model.fit(trainX, trainY, epochs=200, batch_size=32, validation_data=(testX, testY))
```

```python
model.evaluate(valX, valY)

historyData = hist.history

fig = plt.figure(figsize=(10, 5))
ax0 = fig.add_subplot(1,2, 1)
ax0.set_title('Loss')
ax0.grid()
ax0.set_xlabel('Epochs')
ax0.set_ylabel('Loss')
ax0.plot(range(1,len(historyData['loss'])+1), historyData['loss'], c='red', label ='train loss' )
ax0.plot(range(1,len(historyData['val_loss'])+1), historyData['val_loss'], c='green', label ='val loss' )
ax0.legend()

ax1 = fig.add_subplot(1,2, 2)
ax1.set_title('Accuracy')
ax1.grid()
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax1.plot(range(1,len(historyData['accuracy'])+1), historyData['accuracy'], c='red', label ='train accuracy' )
ax1.plot(range(1,len(historyData['val_accuracy'])+1), historyData['val_accuracy'], c='green', label ='val accuracy' )
ax1.legend()
```

### Save model

```python
from tensorflow.keras.optimizers import Adam
# opti = Adam(learning_rate=0.0005, beta_1=0.9, beta_2=0.999)
opti = Adam(learning_rate=0.0005)
```

```python
modelDropoutSaved = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(100, 100, 3)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.1),
  tf.keras.layers.Dense(64, activation=tf.nn.relu),
  tf.keras.layers.Dropout(0.1),
  tf.keras.layers.Dense(NUM_CLASS, activation=tf.nn.softmax)
])
modelDropoutSaved.compile(
    loss='categorical_crossentropy',
    optimizer= opti, 
    metrics=['accuracy'],
    )
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

checkPointCallBack = ModelCheckpoint('checkPointCallBack.h5', save_best_only=True)
earlyStoppingCallBack = EarlyStopping(
    patience=20, monitor='val_loss',
    restore_best_weights=True,
    )
```

```python

histDropoutSaved = modelDropoutSaved.fit(
    trainX, trainY, epochs=200, batch_size=32, 
    validation_data=(testX, testY),
    callbacks=[checkPointCallBack, earlyStoppingCallBack]
    )
```

```python
modelDropoutSaved.evaluate(valX, valY)
historyData = histDropoutSaved.history

fig = plt.figure(figsize=(10, 5))
ax0 = fig.add_subplot(1,2, 1)
ax0.set_title('Loss')
ax0.grid()
ax0.set_xlabel('Epochs')
ax0.set_ylabel('Loss')
ax0.plot(range(1,len(historyData['loss'])+1), historyData['loss'], c='red', label ='train loss' )
ax0.plot(range(1,len(historyData['val_loss'])+1), historyData['val_loss'], c='green', label ='val loss' )
ax0.legend()

ax1 = fig.add_subplot(1,2, 2)
ax1.set_title('Accuracy')
ax1.grid()
ax1.set_xlabel('Epochs')
ax1.set_ylabel('Accuracy')
ax1.plot(range(1,len(historyData['accuracy'])+1), historyData['accuracy'], c='red', label ='train accuracy' )
ax1.plot(range(1,len(historyData['val_accuracy'])+1), historyData['val_accuracy'], c='green', label ='val accuracy' )
ax1.legend()
```

```python
predY = modelDropoutSaved.predict(trainX)
print(predY.shape)
print(predY[0])
```

```python
#  https://stackoverflow.com/questions/47435526/what-is-the-meaning-of-axis-1-in-keras-argmax
from sklearn.metrics import confusion_matrix
from tensorflow.keras.utils import to_categorical

import seaborn as sns
sns.set(style='dark')

yTestYOnehot = trainY

plt.figure(figsize=(12,12))
conMat = confusion_matrix(np.argmax(yTestYOnehot, axis=1), np.argmax(predY, axis=1))
sns.heatmap(conMat, annot=True, cmap='Reds', fmt='d')
plt.xlabel('Pred')
plt.ylabel('Ground Truth')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(np.argmax(yTestYOnehot, axis=-1),  np.argmax(predY, axis=-1)))
```

```python
#    
diff = target - y_pred
diff_idx = []


for i in range(0, len(target)):
    if(diff[i]!=0):
        diff_idx.append(i)
print(len(diff_idx))
print(diff_idx)
```

```python
# visualizing differences
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 15))
for i in range(4*5):
    ax = plt.subplot(4, 5, i+1)
    raw_idx = diff_idx[i]
    ax.title.set_text("t:" + str(classList[target[raw_idx]])  + ', p:' + str(classList[y_pred[raw_idx]]))
    plt.imshow(
        dataArray[raw_idx] / 255,
        # .reshape((28, 28)), 
        cmap='Greys')
plt.show()
```
