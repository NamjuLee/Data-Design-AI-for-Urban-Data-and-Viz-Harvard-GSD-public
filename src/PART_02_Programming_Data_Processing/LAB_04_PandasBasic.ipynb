# Pandas 

```python
# import essential library
import pandas as pd
print(pd.__version__)
```

```python
# pandas data frame from dictionary in python
data = { 
    "name" : ['a','b','c','d','e','f'],
    "id" : [1,2,3,14,15,16],
    "score" : [23,21,18,29,43,21]
}

print (type(data))
print (data)
```

```python
df = pd.DataFrame(data)
print (type(df))
print (df)
```

```python
print(df.columns.values)
print('--------------------------')
print(df.shape)
print('--------------------------')
print(df.head(3))
print('--------------------------')
print(df.tail(2))
print('--------------------------')
print(df.columns.values)
```

```python
### manipulating data frames

theIndex = [2017, 2016, 2015, 2014]
dict1 = {'MA':[80,85,73,99],
       'rate':[5,8,3,5],
       'value':[33,23,42,28]}
dict2 = {'NY':[67,88,76,89],
       'rate':[7,9,6,4],
       'value':[35,53,45,25]}
dict3 = {'CA':[20,43,53,56],
       'rate':[7,4,9,9],
       'value':[13,33,43,38]}
df1 = pd.DataFrame(dict1, index=theIndex)
df2 = pd.DataFrame(dict2, index=theIndex)
df3 = pd.DataFrame(dict3, index=theIndex)
print (df1)
print (df2)
print (df3)
```

```python
concat = pd.concat([df1, df2, df3])
print (concat.shape)
print (concat)
```

```python
df4 = df1.append(df2)
print (df4)
```

```python
# pandas data frame from csv
def GetPandasFromFileCSV(path):
    return pd.read_csv(path, delimiter=',')
def GetPandasFromFileCSVWithoutCol(path):
    return pd.read_csv(path, delimiter=',', index_col=0)

path = "../../public/static/data/Blue_Bike_Stations.csv"
df = GetPandasFromFileCSV(path)
print (df.head(5))
print (df.tail(3))
```

```python
# to List
myList = df['ObjectId'].tolist()
print (type(myList))
print (len(myList))
print ("------------------")
print (myList)
```

```python
import numpy as np
# to Numpy array
npArray = np.array(df['ObjectId'])
print(type(npArray))
print ("------------------")
print(npArray)
```

```python
# to Pandas Data Frame
temp = df[['Latitude', 'Longitude', 'District']]
newNp = np.array(temp)
print(newNp)
print ("------------------")
print(newNp[5:10])
print ("------------------")

newDF = pd.DataFrame(temp)
print(newDF)
```

```python
# rename columns 

print(newDF.head(3))
print ("------------------")
newDF.columns = ["y", "x", 'area']
print(newDF.head(3))
print ("------------------")
newDFRenamed = newDF.rename(columns={"area" : 'city' }) # rename selectively
print(newDFRenamed.head(3))
```

```python
# get data from selected columns
dataCol = newDF['x'].values
print(dataCol)

print('-------------------')
dataCol = newDF['x']
print(dataCol)
```

```python
# slice data between start and end index 
dataCol = newDF['x']
# print dataCol
print(dataCol[5:20])
```

```python
# data change by index
dataCol[9] = 1 
dataCol.iloc[6] = 8888888888 # for 2D dataCol.iloc[:, 9]
print (dataCol[5:12])
```
