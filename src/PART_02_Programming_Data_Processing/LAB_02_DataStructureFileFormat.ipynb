```python
class Person:
  
  def __init__(self, name, address, age, grade):
    self.name = name
    self.address = address
    self.age = age
    self.grade = grade

  def updateGrade(self, addNum):
      for i in range(len(self.grade)):
          self.grade[i] += addNum

p1 = Person("NJ", 21, 'boston', [10, 20, 30])
print(p1.grade)
```

## Queue & Stack

https://computationaldesign.tistory.com/54

```python
# https://github.com/NamjuLee/designComputation/blob/master/Queue.py
import copy
from collections import deque

class Queue:
    data = []
    def __init__(self):
        self.data = []
    def push(self, val):
        if(type(val)==list):
            for i in val: self.data.append(i)
        else: self.data.append(val)
    def pop(self):
        if(len(self.data) == 0 ):
            print ("no data remaining")
            return None
        else:return self.data.pop(0)
    def __eq__(self, other):
        if(len(self.data) != len(other.data)):
            return False
        else:
            num = 0
            for i in range(len(self.data)):
                if(self.data[i] == other.data[i]):
                    num += 1
            if(num == len(self.data)): return True
            else: return False
    def __ne__(self, other):
        if(self == other):return False
        else:return True
    def __str__(self):
        theStr = ""
        for i in self.data:
            theStr += str(i)
        return theStr
```

```python
theQueue = Queue()
theList = [1,2,3,4]
otherList = [1,2,3,4, 5]
num = 5

theQueue.push(theList)
theQueue.push(num)
print (theQueue)

otherQueue = Queue()
otherQueue.push(otherList)
print (otherQueue)

theQueue.pop()
otherQueue.pop()

print (theQueue == otherQueue) 

print (theQueue)
print (otherQueue)
```

```python
# https://github.com/NamjuLee/designComputation/blob/master/Stack.py

import copy
from collections import deque

class Stack:
    data = []
    def __init__(self):
        self.data = []
    def push(self, val):
        if(type(val)==list):
            for i in val:
                self.data.append(i)
        else:
            self.data.append(val)
    def pop(self):
        if(len(self.data) == 0 ):
            print ("no data remaining")
            return None
        else:return self.data.pop(-1)
    def __eq__(self, other):
        if(len(self.data) != len(other.data)):
            return False
        else:
            num = 0
            for i in range(len(self.data)):
                if(self.data[i] == other.data[i]):
                    num += 1
            if(num == len(self.data)): return True
            else: return False
    def __ne__(self, other):
        if(self == other):return False
        else:return True
    def __str__(self):
        theStr = ""
        for i in self.data:
            theStr += str(i)
        return theStr
```

```python
theStack = Stack()
theList = [1,2,3,4]
otherList = [1,2,3,4, 5]
num = 5

theStack.push(theList)
theStack.push(num)
print (theStack)

otherStack = Stack()
otherStack.push(otherList)
print (otherStack)

theStack.pop()
otherStack.pop()

print (theStack == otherStack) 

print (theStack)
print (otherStack)
```

## Graph

https://github.com/NamjuLee/Introduction-to-computational-design-python-notebook

```python
import random
import math

class Node:
    def __init__(self, id, width, height):
#         print('node')
        x = random.random() * width
        y = random.random() * height;
        self.pos = { 'x': x ,'y': y  }
        self.id = id
        self.connections = []
    def addNode(self, other):
        self.connections.append(other);
    def print(self):
        print('id:', self.id, ', x:', self.pos['x'] , ', Y:', self.pos['y'] , ', connection:', self.connections )
```

```python
numOfNodes = 200
nodes = []
for i in range(numOfNodes):
    n = Node(i, 500, 500)
    nodes.append(n);

nodes[10].print()
```

```python
for i in range(300):
    index0 = math.floor( random.random() * numOfNodes )
    index1 = math.floor( random.random() * numOfNodes )
    if (index0 != index1):
        nodes[index0].addNode(nodes[index1])
nodes[2].connections[0].print()
```

```python
for i in nodes:
    i.print()
```

```python
class Graph:
    def __init__(self, num, numOfConnections):
#         print('node')
        self.num = num
        self.width = 500
        self.height = 500
        self.nodes = []
        for i in range(self.num):
            self.nodes.append(Node(i, self.width, self.height))
        self.init(numOfConnections)
    def init(self, num=100):
        for i in range(num):
            index0 = math.floor( random.random() * self.num )
            index1 = math.floor( random.random() * self.num )
            if (index0 != index1):
                self.nodes[index0].addNode(self.nodes[index1])
    def print(self):
        print('numOfnodes:', len(self.nodes))
```

```python
g = Graph(70, 100)
g.print()
g.nodes[50].print()
```

```python
import tkinter
root = tkinter.Tk()
myCanvas = tkinter.Canvas(root, bg="blue", height=g.height, width=g.width)
myCanvas.pack()

def drawPoint(x, y, r = 2): #center coordinates, radius
    myCanvas.create_oval(x - r, y - r, x + r, y + r, outline='white', fill="red", width=2, stipple="gray50")
def drawLine(x0, y0, x1, y1 ):
    myCanvas.create_line(x0, y0, x1, y1, dash=(4, 2), fill="#fb0")

for i in g.nodes:
    x = i.pos['x']
    y = i.pos['y']
    drawPoint(x,y, (len(i.connections)+1 ) * 2)
    
    if len(i.connections) > 0:
        for other in i.connections:
            xOther = other.pos['x']
            yOther = other.pos['y']
            drawLine(x,y, xOther,yOther )
    
root.mainloop()
```
