```python
def OpenTxt(path):
    outTemp = []
    with open(path, 'r') as f: 
        for line in f:
            outTemp.append(line)
    return outTemp

data = OpenTxt('../../public/static/data/Blue_Bike_Stations.csv')
```

```python
# https://jsonlint.com/
import urllib.request, json 
with urllib.request.urlopen("https://raw.githubusercontent.com/NamjuLee/data/master/Boston/MBTA_stop.geojson") as url: 
    data = json.load(url)
    print(data)
```

```python

for i in data["features"]:
    print(i['properties']['area'])
    print(i['geometry']['coordinates'])
```

```python
myData = []

for i in data["features"]:

    area = i['properties']['area']
    pos = i['geometry']['coordinates']

    myData.append(
        {
            "area": area,
            "location": pos
        }
    )
```

```python
myData
```

```python
def SaveJson(path, d):
    with open(path, "w") as writeJSON:
        json.dump(d, writeJSON, indent=4)
```

```python
path = './data/myData.json'
SaveJson(path, myData)
```

```python
def OpenJson(path): 
    json_file = open(path , 'r')
    json_str = json_file.read()
    json_data = json.loads(json_str)
    return json_data
```
