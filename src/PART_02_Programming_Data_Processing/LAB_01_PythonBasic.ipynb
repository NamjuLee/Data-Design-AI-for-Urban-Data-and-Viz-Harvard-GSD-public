```python
def find_A_character_from_word(word):
    theWord = word.upper()
    for i, char in enumerate(theWord):
        if(char == "A"):
            print ("yes")
            return i
    print ("no")
    return -1

theIndex = find_A_character_from_word("apple")
print (theIndex)
print ("--------------------------------")

theIndex = find_A_character_from_word("lemon")
print (theIndex)
print ("--------------------------------")

theIndex = find_A_character_from_word("strawberry")
print (theIndex)
```
